```python
import matplotlib.pyplot as plt
 
# x-axis values
x = [0,1.992,3.988,5.99,7.988,9.880,11.97,13.96,15.96,17.96,19.96,29.98,39.98]
# y-axis values
y = [0,2.76,2.32,2.84,1.64,1.24,1.16,0.8,0.8,0.72,0.8,0.96,0.64]
 
# plotting points as a scatter plot
plt.scatter(x, y, color= "black",
            marker= "o", s=30)
 
# x-axis label
plt.xlabel('Delay Time t (ms)', fontsize=13)
# y-axis label
plt.ylabel('Amplitude A(t) (V)', fontsize=13)
# plot title
plt.title('A(t) vs t', fontsize = 14)
# show legend
#plt.legend()
 
# function to show the plot
plt.show()
```
